```python
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
```

```python
# read all the csv files required
df=pd.read_excel('my_csv/21-22/Player Stats 21-22.xlsx')
creative=pd.read_excel('my_csv/21-22/Creative Magic 21-22.xlsx')
vision=pd.read_excel('my_csv/21-22/Pass Types 21-22.xlsx')
pass_stats=pd.read_excel('my_csv/21-22/Passing Stats 21-22.xlsx')
pose=pd.read_excel('my_csv/21-22/Player Possession 21-22.xlsx')
stats=pd.read_excel('my_csv/21-22/Player Stats 21-22.xlsx')
```

```python
# declare mp equal to the number of 90 minutes played by a player
mp=df['90s']
#declare temp equal to the pass completion rate of a player
temp=vision['Cmp']/vision['Att']

#dataframe - creative equal to the sum of the goal creation and shot creation chances per 90 of each player
creative=creative['GCA90'].add(creative['SCA90'])

#dataframe - vision with column values equal to the number of through balls and completion rate
vision=vision[['TB']]
vision['Comprate']=temp

#dataframe - pass_stats with column values equal to the number of passes in the box completed, crosses in the box completed, progressive passses completed, passes completed in the final third
pass_stats=pass_stats[['PPA','CrsPA','Prog','1/3rd']]

#declare drib equal to the dribble rate of a player
drib=pose['Succ']/pose['Att']
# converting drib to a dataframe from a series
drib=drib.to_frame()
```

```python
#divide stats by number of 90s played wherever required
vision['TB']=vision['TB'].div(mp)

pass_stats['PPA']=pass_stats['PPA'].div(mp)
pass_stats['CrsPA']=pass_stats['CrsPA'].div(mp)
pass_stats['Prog']=pass_stats['Prog'].div(mp)
pass_stats['1/3rd']=pass_stats['1/3rd'].div(mp)
```

```python
#dropping unnecessary columns from 'df' dataframe
df=df.drop(columns=['Rk','Nation','Age','Born'])
df=df.drop(df.iloc[:,9:],axis=1)
#make new columns in df, and assign them their respective values
df=df.assign(SGCA90=creative,through_ball=vision['TB'],complete_rate=vision['Comprate'],pass_in_the_box=pass_stats['PPA'],crosses_in_the_box=pass_stats['CrsPA'],progressive_passes=pass_stats['Prog'],pass_in_last_third=pass_stats['1/3rd'],dribble_rate=drib)
#fill entries containing NaN with 0
df=df.fillna(0)
# filter df such that only players who have played more than 5 90s are displayed
df=df.loc[df['90s']>5]
```

```python
# set a variable that is equalt to the correlation of columns with each other
correlate=df.corr()
# setting seperate variable equal to the correlation coefficient of SGCA90 and different stats
assist_coeff=correlate['SGCA90']['Ast']
through_coeff=correlate['SGCA90']['through_ball']
complete_coeff=correlate['SGCA90']['complete_rate']
box_pass_coeff=correlate['SGCA90']['pass_in_the_box']
cross_coeff=correlate['SGCA90']['crosses_in_the_box']
prog_pass_coeff=correlate['SGCA90']['progressive_passes']
last_third_coeff=correlate['SGCA90']['pass_in_last_third']
drib_coeff=correlate['SGCA90']['dribble_rate']
```

```python
# creative index is equal to the sum of the product of all stats and their respective correlation coefficient
df['creative_index']=assist_coeff*df['Ast']+through_coeff*df['through_ball']+complete_coeff*df['complete_rate']+box_pass_coeff*df['pass_in_the_box']+cross_coeff*df['crosses_in_the_box']+prog_pass_coeff*df['progressive_passes']+last_third_coeff*df['pass_in_last_third']+drib_coeff*df['dribble_rate']
# sort dataframe in descending order according to creative index value
df.sort_values(by='creative_index',ascending=False).head(10).reset_index(drop=True,)
```
